```python
import pandas as pd 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
df = pd.read_csv("iris.csv") 
 
(train_set, test_set) = train_test_split(df.values, train_size=0.7, 
random_state=278823)

train_data, train_target = train_set[:, :-1], train_set[:, -1]
test_data, test_target = test_set[:, :-1], test_set[:, -1]
```

```python
knn = KNeighborsClassifier(n_neighbors=3, metric='euclidean')
knn.fit(train_data, train_target)
knn.score(test_data, test_target)
```

```python
predictions = knn.predict(test_data)
cm = confusion_matrix(test_target, predictions)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=knn.classes_)
disp.plot()
```

```python
knn = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn.fit(train_data, train_target)
knn.score(test_data, test_target)
```

```python
predictions = knn.predict(test_data)
cm = confusion_matrix(test_target, predictions)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=knn.classes_)
disp.plot()
```

```python
knn = KNeighborsClassifier(n_neighbors=11, metric='euclidean')
knn.fit(train_data, train_target)
knn.score(test_data, test_target)
```

```python
predictions = knn.predict(test_data)
cm = confusion_matrix(test_target, predictions)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=knn.classes_)
disp.plot()
```

```python
model = GaussianNB()
model.fit(train_data, train_target)
model.score(test_data, test_target)
```

```python
predictions = model.predict(test_data)
cm = confusion_matrix(test_target, predictions)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot()
```

Największą precyzję otrzymaliśmy dla klasyfikatora k-najbliższych sąsiadów z parametrem k równym 5 lub 11 (97,78%)
