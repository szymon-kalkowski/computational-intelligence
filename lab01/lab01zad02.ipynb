```python
import pandas
import matplotlib as plt
from matplotlib import pyplot

miasta = pandas.read_csv('miasta.csv')

print(miasta.values)
print("")

miasta.loc[len(miasta)] = [2010, 460, 555, 405]
print(miasta.values)
```

```python
pyplot.plot(miasta['Rok'], miasta['Gdansk'], 'ro-', label='Gdansk')
pyplot.xlabel('Lata')
pyplot.ylabel('Liczba ludnosci [w tys.]')
pyplot.title('Ludnosc w miastach Polski')
pyplot.plot(miasta['Rok'], miasta['Poznan'], 'bo-', label='Poznan')
pyplot.plot(miasta['Rok'], miasta['Szczecin'], 'go-', label='Szczecin')
pyplot.legend()
```

```python
standGdansk = (miasta['Gdansk'] - miasta['Gdansk'].mean()) / miasta['Gdansk'].std()
standPoznan = (miasta['Poznan'] - miasta['Poznan'].mean()) / miasta['Poznan'].std()
standSzczecin = (miasta['Szczecin'] - miasta['Szczecin'].mean()) / miasta['Szczecin'].std()

print("Srednia i odchylenie standardowe dla Gdańska:")
print(standGdansk.mean())
print(standGdansk.std())
print("")

print("Srednia i odchylenie standardowe dla Poznania:")
print(standPoznan.mean())
print(standPoznan.std())
print("")

print("Srednia i odchylenie standardowe dla Szczecina:")
print(standSzczecin.mean())
print(standSzczecin.std())
print("")
```

```python
normGdansk = (miasta['Gdansk'] - miasta['Gdansk'].min()) / (miasta['Gdansk'].max() - miasta['Gdansk'].min())
normPoznan = (miasta['Poznan'] - miasta['Poznan'].min()) / (miasta['Poznan'].max() - miasta['Poznan'].min())
normSzczecin = (miasta['Szczecin'] - miasta['Szczecin'].min()) / (miasta['Szczecin'].max() - miasta['Szczecin'].min())

print("Min i max dla Gdańska:")
print(normGdansk.min())
print(normGdansk.max())
print("")

print("Min i max dla Poznania:")
print(normPoznan.min())
print(normPoznan.max())
print("")

print("Min i max dla Szczecina:")
print(normSzczecin.min())
print(normSzczecin.max())
print("")
```
